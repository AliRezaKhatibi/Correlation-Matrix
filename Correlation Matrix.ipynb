```python
from numpy import mean ,sqrt ,array 
from numpy.linalg import det
import numpy as np
##########################
import seaborn as sns
import matplotlib.pyplot as plt
```

### 1. Mathematical Expectation (Expectation)
The mathematical expectation, or mean, of a random variable is generally considered the average value of the random variable.

The formula for the mathematical expectation of a random variable X is:
\\[
E(X) = \sum_{i} x_i \cdot P(x_i)
\\]
Where:
- \\(E(X)\\) is the mathematical expectation of the random variable X.
- \\(x_i\\) are the possible values of X.
- \\(P(x_i)\\) is the probability of each of these values.

### 2. Variance
Variance of a random variable measures the spread or dispersion of the values of the variable around the expectation.

The formula for variance is:
\\[
Var(X) = E(X^2) - (E(X))^2
\\]
Where:
- \\(Var(X)\\) is the variance of the random variable X.
- \\(E(X^2)\\) is the expectation of the squared values of the variable.
- \\((E(X))^2\\) is the square of the expectation of the variable.

### 3. Standard Deviation
Standard deviation is the square root of variance and represents the spread of values around the mean.

The formula for standard deviation is:
\\[
\sigma_X = \sqrt{Var(X)}
\\]
Where:
- \\(\sigma_X\\) is the standard deviation of the random variable X.
- \\(Var(X)\\) is the variance of the random variable X.

### 4. Covariance
Covariance is a measure of the linear relationship between two random variables. If the covariance is positive, the variables change together in the same direction, and if negative, they change in opposite directions.

The formula for covariance is:
\\[
Cov(X, Y) = E[(X - E(X)) \cdot (Y - E(Y))]
\\]
Where:
- \\(Cov(X, Y)\\) is the covariance between two random variables X and Y.
- \\(E(X)\\) and \\(E(Y)\\) are the expectations of X and Y, respectively.
### Covariance (Alternative Form)

Covariance can also be written as:

\\[
Cov(X, Y) = E(XY) - E(X) \cdot E(Y)
\\]

Where:
- \\(Cov(X, Y)\\) is the covariance between two random variables X and Y.
- \\(E(XY)\\) is the expectation of the product of X and Y (i.e., the expected value of the product of the two random variables).
- \\(E(X)\\) is the expectation (mean) of X.
- \\(E(Y)\\) is the expectation (mean) of Y.

This form highlights that covariance is the difference between the expected value of the product of X and Y, and the product of their individual expectations.


### 5. Correlation
Correlation is a standardized version of covariance, measuring the strength and direction of the linear relationship between two variables. It ranges from -1 (perfect negative correlation) to 1 (perfect positive correlation), with 0 indicating no linear relationship.

The formula for correlation is:
\\[
\rho_{XY} = \frac{Cov(X, Y)}{\sigma_X \cdot \sigma_Y}
\\]
Where:
- \\(\rho_{XY}\\) is the correlation coefficient between X and Y.
- \\(Cov(X, Y)\\) is the covariance between X and Y.
- \\(\sigma_X\\) and \\(\sigma_Y\\) are the standard deviations of X and Y, respectively.
### 6. Correlation Matrix

The correlation matrix is a square matrix that contains the correlation coefficients between every pair of variables. If we have multiple random variables \\(X_1, X_2, \dots, X_n\\), the correlation matrix \\(R\\) is calculated as follows:

\\[
R = \begin{bmatrix}
    1 & \rho_{X_1, X_2} & \rho_{X_1, X_3} & \cdots & \rho_{X_1, X_n} \\
    \rho_{X_2, X_1} & 1 & \rho_{X_2, X_3} & \cdots & \rho_{X_2, X_n} \\
    \vdots & \vdots & \vdots & \ddots & \vdots \\
    \rho_{X_n, X_1} & \rho_{X_n, X_2} & \rho_{X_n, X_3} & \cdots & 1
\end{bmatrix}
\\]

Where:
- \\( \rho_{X_i, X_j} \\) is the correlation coefficient between variables \\(X_i\\) and \\(X_j\\), which can be calculated as:

\\[
\rho_{X_i, X_j} = \frac{Cov(X_i, X_j)}{\sigma_{X_i} \cdot \sigma_{X_j}}
\\]

Where:
- \\(Cov(X_i, X_j)\\) is the covariance between \\(X_i\\) and \\(X_j\\).
- \\(\sigma_{X_i}\\) and \\(\sigma_{X_j}\\) are the standard deviations of \\(X_i\\) and \\(X_j\\), respectively.

### Steps to calculate the correlation matrix:
1. Calculate the covariance between each pair of variables.
2. Normalize the covariance by dividing by the standard deviations of the corresponding variables.
3. Fill the matrix with the correlation coefficients.

The resulting matrix will be symmetric, with 1s on the diagonal, as each variable is perfectly correlated with itself.
### 7. Analysis of the Determinant of the Correlation Matrix

- **Determinant = 1 (Perfect Positive Correlation or Linear Independence)**:
  - All variables are perfectly   relatively independent.
  
- **Determinant = 0 (Linear Dependence or Multicollinearity)**:
  - There is linear dependence among variables, meaning at least one variable can be expressed as a linear combination of the others. This leads to multicollinearity and makes it hard to separate individual effects in regression models.

- **Determinant Between 0 and 1 (Partial Correlation and Partial Independence)**:
  - Variables are not perfectly correlated but still show some linear relationships. A smaller determinant (closer to 0) indicates stronger relationships, while a larger determinant (closer to 1) suggests weaker dependencies.

### Key Insights:
- **High determinant** (close to 1) means variables are relatively independent.
- **Low determinant** (close to 0) indicates high correlation and potential multicollinearity issues.
- **Determinant = 0** means perfect linear dependence among variables, making the matrix singular.


```python
# Three sample vectors
x = array([1, 0, 0, 0, 0])  # First vector
y = array([0, 1, 0, 0, 0])  # Second vector
z = array([0, 0, 1, 0, 0])
# z = 2*x - y  # Third vector as a linear combination of x and y

# If we want to add noise, we can activate this line
# z = 2*x - y + np.random.normal(50, 5, size=x.shape)  # Adding noise

# Compute mean values of vectors
E_x = mean(x)  # Mean of x
E_y = mean(y)  # Mean of y
E_z = mean(z)  # Mean of z

# Compute element-wise mean products
E_xy = mean(x * y)  # Mean product of x and y
E_xz = mean(x * z)  # Mean product of x and z
E_yz = mean(y * z)  # Mean product of y and z

# Compute variances
# Instead of using np.var, we calculate variances manually
var_x = mean(x**2) - (mean(x))**2  # Variance of x
var_y = mean(y**2) - (mean(y))**2  # Variance of y
var_z = mean(z**2) - (mean(z))**2  # Variance of z

# Compute covariances
cov_xy = E_xy - E_x * E_y  # Covariance of x and y
cov_xz = E_xz - E_x * E_z  # Covariance of x and z
cov_yz = E_yz - E_y * E_z  # Covariance of y and z

# Compute standard deviations (sigma) for each vector
sigma_x = sqrt(var_x)  # Standard deviation of x
sigma_y = sqrt(var_y)  # Standard deviation of y
sigma_z = sqrt(var_z)  # Standard deviation of z

# Compute correlation coefficients between vectors
corr_xy = cov_xy / (sigma_x * sigma_y)  # Correlation coefficient of x and y
corr_xz = cov_xz / (sigma_x * sigma_z)  # Correlation coefficient of x and z
corr_yz = cov_yz / (sigma_y * sigma_z)  # Correlation coefficient of y and z

# Construct the correlation matrix
R = array([[1, corr_xy, corr_xz],
           [corr_xy, 1, corr_yz],
           [corr_xz, corr_yz, 1]])

# Compute determinant of the correlation matrix
det_R = det(R)

# Print correlation matrix
print("Correlation Matrix (R):")
print(R)

# Print determinant of the correlation matrix
print("Determinant of the correlation matrix:", np.round(det_R, 4))
```

```python
# Print determinant of the correlation matrix
sns.heatmap(R, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
def compute_correlation_matrix(*vectors):
    """
    Compute the correlation matrix for the given input vectors.
    :param vectors: Multiple input vectors
    :return: Correlation matrix
    """
    matrix = np.column_stack((vectors))  # Create a matrix from the input vectors
    correlation_matrix = np.corrcoef(matrix, rowvar=False)  # Compute correlation matrix
    return correlation_matrix




def determinant(matrix):
    """
    Compute the determinant of a matrix using NumPy.
    :param matrix: Square matrix
    :return: Determinant of the matrix
    """
    return np.linalg.det(matrix)
```

```python
# Example
v1 = np.array([0, 0, 1, 0, 0])
v2 = np.array([0, 0, 0, 1, 0])
v3 = np.array([0, 0, 0, 0, 1])
#v3 = 2*v1 - v2
#v3 = 2*v1 - v2 + np.random.normal(10, 5, size=v1.shape)  # Adding noise 



correlation_matrix = compute_correlation_matrix(v1, v2, v3)
det = determinant(correlation_matrix)

print("Correlation Matrix:")
print(correlation_matrix)
print("Determinant of the Correlation Matrix:", round(det, 4))
# Print determinant of the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title("Correlation Matrix Heatmap")
plt.show()
```
